```python
import numpy as np
import itertools
import random


def permute_list(input_list, permutation):
    if len(input_list) != 5 or len(permutation) != 5:
        raise ValueError("Both input_list and permutation should have exactly 5 elements")

    input_array = np.array(input_list)
    permutation_array = np.array(permutation)

    if not np.all(np.sort(permutation_array) == np.arange(5)):
        raise ValueError("Permutation should contain integers from 0 to 4, inclusive")

    output_array = input_array[permutation_array]
    return output_array.tolist()

def is_even_permutation(lst):
    """
    Check if a given list is an even permutation of the numbers from 0 to 4 inclusive.
    
    Args:
    - lst: a list of integers representing a permutation of numbers from 0 to 4 inclusive
    
    Returns:
    - True if the given permutation is even, False otherwise
    """
    if len(lst) != 5 or set(lst) != set(range(5)):
        return False
    
    swaps = 0
    for i in range(4):
        for j in range(i+1, 5):
            if lst[i] > lst[j]:
                swaps += 1
    
    return swaps % 2 == 0

def even_permutations_of_length_5():
    elements = [0, 1, 2, 3, 4]
    even_permutations = []
    for perm in itertools.permutations(elements):
        if is_even_permutation(list(perm)):
            even_permutations.append(list(perm))
    return even_permutations

def random_perm_generator(even = True):
    ls = list(range(5))
    random.shuffle(ls)
    if even:
        while not is_even_permutation(ls):
            random.shuffle(ls)
    return ls

all_perms = even_permutations_of_length_5()
```

```python
def random_group_operation():
    a = random_perm_generator()
    b = random_perm_generator()
    c = permute_list(b,a)

    return f'{all_perms.index(a)} {all_perms.index(b)} {all_perms.index(c)}'

```

```python
def generate_S5_with_identity():
    out = []
    for i in range(60,120):
        out.append(f'0 {i} {i}')
        out.append(f'{i} 0 {i}')
    return out
```

```python
def generate_A5_data(length=50):
    out = []
    while len(out)<length:
        datum = random_group_operation()
        if datum not in out:
            out.append(datum)
    return out
generate_A5_data(5)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python

def create_shape_image():
    # Create an empty image array (height x width x RGB)
    height, width = 100, 200
    img = np.zeros((height, width, 3), dtype=np.uint8)

    # Define red square coordinates
    square_size = 40
    square_top_left = (30, 30)
    square_bottom_right = (square_top_left[0] + square_size, square_top_left[1] + square_size)

    # Fill the red square
    img[square_top_left[0]:square_bottom_right[0], square_top_left[1]:square_bottom_right[1]] = [255, 0, 0]

    # Define blue triangle coordinates
    triangle_vertices = np.array([(150, 30), (190, 70), (150, 70)], dtype=np.int32)

    # Create a mask for the triangle
    triangle_mask = np.zeros((height, width), dtype=np.uint8)
    cv2.fillPoly(triangle_mask, [triangle_vertices], 1)

    # Fill the blue triangle
    img[triangle_mask == 1] = [0, 0, 255]

    # Display the image using imshow
    plt.imshow(img)
    plt.axis('off')
    plt.show()

create_shape_image()
```
